```python
import pandas as pd

# Create two sample DataFrames
data1 = {
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Department': ['HR', 'IT', 'Finance', 'IT']
}

data2 = {
    'EmployeeID': [2, 4, 3, 5],
    'Salary': [70000, 80000, 60000, 90000]
}

# Convert dictionaries to DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Display the original DataFrames
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Merge the DataFrames on 'EmployeeID' using an inner join
merged_df = pd.merge(df1, df2, on='EmployeeID', how='inner')



```

```python
merged_df1 = pd.merge(df1, df2, on='EmployeeID', how='left')
merged_df1
```

```python
merged_df2 = pd.merge(df1, df2, on='EmployeeID', how='right')
merged_df2
```

```python
merged_df3 = pd.merge(df1, df2, on='EmployeeID', how='outer')
merged_df3
```

```python
merged_df3 = pd.merge(df1, df2, how='cross')
merged_df3
```
