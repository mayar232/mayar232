```python
x = df["Imperfections"]
y = df["Repainted Parts"]

sizes = df["Price"] / 70000
colors = df["Price"]  

plt.scatter(x, y,
                 s=sizes,
                 c=colors,
                 cmap="viridis",
                 marker="*")

plt.xlabel("Imperfections")
plt.ylabel("Repainted Parts")
plt.title("Scatter Plot with Price-based Sizes")

plt.colorbar()
plt.show()
```

```python
counts = df["Transmission"].value_counts().reset_index()

plt.pie(
    counts["count"],
    labels=counts["Transmission"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Transmission Distribution")
plt.show()


```

```python
import matplotlib.pyplot as plt

plt.boxplot(
    df[["Price", "Engine capacity"]].values,
    labels=["Price", "Engine capacity"],
    whis=[0, 100],
    patch_artist=True,
    showmeans=True,
    boxprops=dict(facecolor="orange"),
    medianprops=dict(color="maroon"),
    capprops=dict(color="black")
)

plt.yscale("log")
plt.title("Boxplot with Log Scale")
plt.show()
```
