```python
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([15, 15, 35])
print("s1>s2:")
print(s1>s2)
print("s1=s2:")
print(s1==s2)
```

```python
sales = pd.Series([200, 450, 300, 150, 500],
                  index=["Mon", "Tue", "Wed", "Thu", "Fri"])

# Total sales
print("Total:", sales.sum())

# Average sales
print("Average:", sales.mean())

# Max sales day
print("Highest day:", sales.idxmax())

# Sales > 300
print("Days with sales > 300:")
print(sales[sales > 300])
```
