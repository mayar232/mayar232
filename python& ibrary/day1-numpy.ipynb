```python
grid=np.arange(16).reshape((4,4))
print(grid)
print("\n")
upper,lower=np.vsplit(grid,[1])#mean row split from 0 to 1 and the rest is in the lower
print(upper)
print("\n")
print(lower)

```

```python
grid=np.arange(1,17).reshape((4,4))
a,b=np.hsplit(grid,[3])
print(a)
print("--------")
print(b)
```

```python
grid=np.arange(20).reshape((4,5))
a,b=np.hsplit(grid,[2])
print(a)
print("--------")
print(b)
```

```python
grid=np.arange(20).reshape((4,5))
a,b,z=np.hsplit(grid,[2,4])# h mean coloum split from 0 to 1 then 2to 3 and the rest in z
print(a)
print("--------")
print(b)
print("--------")
print(z)
```
