```python
import pandas as pd
```

```python
# Sample Data
# -----------------------------
data = {
    "Date": [
        "2023-01-01", "2023-01-15", "2023-02-10",
        "2023-02-20", "2023-03-05", "2023-03-25"
    ],
    "Sales": [100, 150, 200, 180, 220, 300]
}

df = pd.DataFrame(data)
```

```python
df["Date"]=pd.to_datetime(df["Date"])
print(df.info())
```

```python
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['Weekday']=df['Date'].dt.day_name()
print("\nDate with extracted features:")
print(df)
```

```python
print("\n the filtering in 2023")
print(df[(df["Date"]>= "2023-01-01") & (df["Date"]<="2023-02-28")])
```

```python
marks=[50,60,70,70,80,90,90,100]
plt.hist(marks)
plt.xlabel("Marks")
plt.ylabel("Number Of Student")
plt.title("Histogram Of Mark")
plt.show()
```

```python
marks=[50,60,70,70,80,90,90,100]
plt.hist(marks,
         bins=4,
         color="skyblue",
         edgecolor="black"
        )
         
plt.xlabel("Marks")
plt.ylabel("Number Of Student")
plt.title("Histogram Of Mark")
plt.show()
```

```python
marks=[50,60,70,70,80,90,90,100]
plt.hist(marks,
         bins=5,
         range=(50,100),
         density=True,
         
         color="red",
         alpha=0.6,
         edgecolor="black",
         width=2.9
        )
         
plt.xlabel("Marks")
plt.ylabel("probability density")
plt.title("Histogram  range")
plt.show()
```

```python
data = {
    "product": ["A", "B", "C", "D"],
    "Sales": [100, 150, 200, 180]
}

df = pd.DataFrame(data)

explode = [0.1, 0, 0, 0]

plt.pie(
    df["Sales"],
    labels=df["product"],
    colors=["gold", "skyblue", "pink", "salmon"],
    explode=explode,
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    wedgeprops={"edgecolor": "black"}
)

plt.title("Sales Distribution")
plt.show()
```
