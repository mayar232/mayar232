```python
import numpy as np
```

```python
pationId=np.array([101,102,103,104,105])
heartRate=np.array([72,110,95,130,88])
statue=np.where(heartRate>100 ,"critical","stable")# patiom
print("pation ID: ",pationId)
print("heart rates: ",heartRate)
print("statues: ",statue)
```

```python
#temppp filling non=0
tempreture=np.array([37.2,36.5,0,38.1,0,36.8])
condition=tempreture!=0
filterd=tempreture[condition]
avgTemp=filterd.mean()
print(avgTemp)
cleanedTemb=np.where(tempreture==0,avgTemp,tempreture)
print("original: ",tempreture)
print("cleaned: ",cleanedTemb )
                    
```

```python
condition2 = cleanedTemb > 37# number grate then 37
condition2
filterd =cleanedTemb[condition2]
filterd
```

```python
ages=np.array([23,25,26,70,50])
bp=np.array([120,150,135,160,145])
high= np.where((ages>60)|(bp>140),"high", "normal")
print("the age: ",ages)
print(" the blood pressure: ",bp)
print(high)
```

```python
salaris=np.array([400,600,550,300,700])
condition=salaris>=500
bones=np.where(condition, salaris+100,salaris+50)
print(bones)
```

```python
marks=np.array([90,75,60,82,90,99,95,86])
grades=np.where(marks>=85,np.where(marks>=95,"A+","A"),np.where(marks>=70,"B","C"))
print(marks)
print(grades)
```

```python
purches=np.array([50,200,40,35,100,120])
discount=np.where(purches>100,purches*0.9,purches*0.8)
print(discount)
```

```python
heartRate = np.array([80, 130, 110, 120, 70])
oxygen = np.array([95, 80, 92, 88, 70])

emergency = np.where((heartRate > 120) & (oxygen < 90), "emergency", "normal")

print("heart Rate:", heartRate)
print("oxygen level:", oxygen)
print(emergency)
```
