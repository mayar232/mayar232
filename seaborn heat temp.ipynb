```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame (example: student performance dataset)
data = {
    "StudyHours": [2, 3, 4, 5, 6],
    "Marks": [50, 60, 70, 80, 90],
    "Attendance": [60, 65, 70, 85, 95],
    "SleepHours": [8, 7, 6, 6, 5]
}
df = pd.DataFrame(data)
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(
    corr,annot=True,
    cmap="coolwarm",
    vmax=1,vmin=-1,
    linewidths=0.5)
plt.title("correlation Heat Map")
plt.show()
```
